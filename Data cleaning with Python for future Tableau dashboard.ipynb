### 1.3. Data cleaning

#### Irrelevant columns

Let's drop the column "Column1" (last column of the dataset, after "Revenue"):

```python
df = df.drop(columns=["Column1"])
df
```

#### Customer Age

"Customer Age" is a "float" Datatype: let's convert it into an "integer" datatype. 
First let's make sure there are no  NaN (not a number) values in the whole dataset:

```python
df.isnull().sum()
```

```python
df.isnull()
```

Row 34866 has NaN cells. Let's take a closer look at this row:

```python
df.loc[[34866]]
```

```python
# Let's drop row 34866:
df = df.drop(labels=[34866], axis=0)
```

```python
df.isnull().sum()
```

Now we can convert "Customer Age" from "float" into "int":

```python
df['Customer Age'] = df['Customer Age'].astype('int')
```

```python
df.info()
```

#### Customer Gender

Since we don't want to run a linear regression, we don't need to convert customer gender into a dummy variable. But let's remap the values from "M/F" into "male/female":

```python
df['Customer Gender'] = df["Customer Gender"].map({'M': "Male",'F': "Female"})
```

```python
df
```

#### Date

Let's convert "Date" from "string/object" to "datetime":

```python
df.Date[0]
```

```python
df.Date = pd.to_datetime(df['Date'])
```

```python
df.info()
```

#### Year

Let's convert "Year" from "float" to "int32":

```python
df.Year = df.Year.astype('int32')
```

```python
df.info()
```

Let's convert "Quantity" from "float" to "int32":

```python
df[["Quantity"]] = df[["Quantity"]].astype('int32')
```

```python
df.info()
```

#### New column: Unit Margin

Let's create a new column "Unit Margin", which is the difference between the original production cost of a product and its final price:

```python
df['Unit Margin'] = df['Unit Price'] - df['Unit Cost']
```

```python
df
```

#### New column: Profit

Let's create a new column "Profit", which will show  the actual profit for each purchase (Revenue - Cost):

```python
df['Profit'] = df['Revenue'] - df['Cost']
df
```

#### Changing the place of "Unit Margin" column

Let's shift the new column "Unit Margin" after "Unit Price": first we bring all the columns into a list 

```python
columns = df.columns.tolist()
columns
```

```python
df = df[['index',
 'Date',
 'Year',
 'Month',
 'Customer Age',
 'Customer Gender',
 'Country',
 'State',
 'Product Category',
 'Sub Category',
 'Quantity',
 'Unit Cost',
 'Unit Price',
 'Unit Margin',
 'Cost',
 'Revenue',
 'Profit']]
```

```python
df
```

#### "Float" variables from six decimals to two decimals

Let's reduce "Unit Cost", "Unit Price", "Unit Margin", "Cost", "Revenue", "Profit" to 2 decimals:

```python
df= df.round(decimals=2)
df
```

### Saving the new dataset in a new CSV-File

Now the dataset is ready to be used for data visualisation purposes. Let's save the cleaned dataset under a new CSV file:

```python
df.to_csv("bike_store_data_for_dashboard.csv", date_format='%Y-%m-%d', index= False)
```
